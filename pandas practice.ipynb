# GROUP BY

The Groupby method is used to group rows together and perform aggregate functions

```python
# create dataframe as given below
data = {'custID':['1001','1001','1002','1002','1003','1003'],
        'custName':['UIPat','DatRob','Goog','chrysler','Ford','GM'],
       'Profitinlakhs':[2005,3245,1245,8765,5463,3547]}
```

```python
dataframe = pd.DataFrame(data)
```

```python
dataframe
```

we can now use the groupby method to group rows together based on a cloumn name. for example lets group based on custID. This will create a DataFramegruopby object.

```python
dataframe.groupby('custID')
```

this object can be saved as a variable

```python
custID_grouped = dataframe.groupby('custID')
```

```python
custID_grouped.mean()
```

```python
dataframe.groupby('custID').median
```

```python
dataframe.groupby('custID').median()
```

```python
custID_grouped.std()
```

```python
custID_grouped.mean()
```

```python
custID_grouped.max()
```

```python
dataframe.groupby('custName').mean()
```

# Joining

join can be used to combine columns of 2 dataframes that have different index values into a singal dataframe

The one difference between merge and join is that, merge uses comman columns to combine two dataframes, whereas join uses the row index to join two dataframes

```python
table1 = pd.DataFrame({'Q1':['101','1006','1003','1004'],
                      'Q2':['201','202','203','204']})
table2 = pd.DataFrame({'Q3':['301','302','303','304'],
                      'Q4':['401','402','403','404']})
```

```python
table1.join(table2)
```

```python
table1.join(table2,how='outer')
```
